```
import GetOldTweets3 as got
import pandas as pd
import time
username='realdonaldtrump'
limit=10000
print(time.asctime(time.localtime(time.time())))
tweetCriteria = got.manager.TweetCriteria().setUsername(username).setTopTweets(True).setMaxTweets(limit)
# Creation of list that contains all tweets
tweets = got.manager.TweetManager.getTweets(tweetCriteria)
# Creating list of chosen tweet data
print(time.asctime(time.localtime(time.time())))
trump_tweets = [[tweet.username, tweet.text,tweet.favorites,tweet.retweets,tweet.replies,tweet.date] for tweet in tweets]
# Creation of dataframe from tweets list
tweets_df = pd.DataFrame(trump_tweets)
```

```
tweets_df.columns=['username','tweet','likes','retweets','comments','date']
```

```
tweets_df['username'].unique()
```

```
tweets_df
```

```
type(tweets_df['date'][0])
```

```
dates=tweets_df['date'].astype(str)
```

```
month=[]
year=[]
day=[]
hour=[]
for date in dates:
  year.append(date[:4])
  month.append(date[5:7])
  day.append(date[8:10])
  hour.append(date[11:13])
```

```
len(month)
```

```
month=pd.Series(month)
year=pd.Series(year)
day=pd.Series(day)
hour=pd.Series(hour)
```

```
tweets_df.drop(['username','date'],axis=1,inplace=True)
```

```
preprocessed_tweets_df=pd.concat([tweets_df,year,month,day,hour],axis=1)
```

```
preprocessed_tweets_df.columns=['tweet','likes','retweets','comments','year','month','day','hour']
```

```
preprocessed_tweets_df
```

```
length_of_tweets=[]
for tweet in preprocessed_tweets_df['tweet']:
  length_of_tweets.append(len(tweet))
```

```
length_of_tweets=pd.Series(length_of_tweets)
preprocessed_tweets_df=pd.concat([preprocessed_tweets_df,length_of_tweets],axis=1)
preprocessed_tweets_df.columns=['tweet','likes','retweets','comments','year','month','day','hour','length']
```

```
preprocessed_tweets_df.head()
```

```
preprocessed_tweets_df['year']=preprocessed_tweets_df['year'].astype(int)
preprocessed_tweets_df['month']=preprocessed_tweets_df['month'].astype(int)
preprocessed_tweets_df['day']=preprocessed_tweets_df['day'].astype(int)
preprocessed_tweets_df['hour']=preprocessed_tweets_df['hour'].astype(int)
```

Lets preprocess the data more exactly

```
time_of_day=[]
for h in preprocessed_tweets_df['hour']:
  if h<12 and h>=6:
    time_of_day.append('morning')
  elif h>=12 and h<18:
    time_of_day.append('day')
  elif h>=18:
    time_of_day.append('evening')
  else:
    time_of_day.append('night')
```

```
time_of_day=pd.Series(time_of_day)
```

```
time_of_day.unique()
```

```
season=[]
for m in preprocessed_tweets_df['month']:
  if m<=2 and m>=1:
    season.append('winter')
  elif m>=3 and m<=5:
    season.append('spring')
  elif m>=6 and m<=8:
    season.append('summer')
  elif m>=9 and m<=11:
    season.append('autumn')
  else:
    season.append('winter')
```

```
season=pd.Series(season)
```

```
preprocessed_tweets_df=pd.concat([preprocessed_tweets_df,season,time_of_day],axis=1)
```

```
preprocessed_tweets_df.columns=['tweet','likes','retweets','comments','year','month','day','hour','length','season','time_of_day']
preprocessed_tweets_df
```

```
import plotly.express as px
import numpy as np
df = preprocessed_tweets_df
fig = px.scatter_3d(df, x='season', y='time_of_day', z='year', size='likes', color='retweets')
fig.update_layout(scene_zaxis_type="log")
fig.show()
```

```
import plotly.graph_objects as go
seasons=list(df['season'].unique())

likes_for_summer=df[df['season']=='summer']['likes'].sum()
likes_for_spring=df[df['season']=='spring']['likes'].sum()
likes_for_winter=df[df['season']=='winter']['likes'].sum()
likes_for_autumn=df[df['season']=='autumn']['likes'].sum()
comments_for_summer=df[df['season']=='summer']['comments'].sum()
comments_for_spring=df[df['season']=='spring']['comments'].sum()
comments_for_winter=df[df['season']=='winter']['comments'].sum()
comments_for_autumn=df[df['season']=='autumn']['comments'].sum()

fig = go.Figure(data=[
    go.Bar(name='likes', x=seasons, y=[likes_for_summer,likes_for_spring,likes_for_winter,likes_for_autumn]),
    go.Bar(name='comments', x=seasons, y=[comments_for_summer,comments_for_spring,comments_for_winter,comments_for_autumn])
])
# Change the bar mode
fig.update_layout(barmode='group')
fig.show()
```
